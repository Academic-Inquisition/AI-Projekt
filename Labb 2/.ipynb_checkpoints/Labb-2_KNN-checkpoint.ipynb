```python
# Imports
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read our Data CSV's
df = pd.read_csv('data/datatraining.csv', delimiter=';')
print(df.head())
```

```python
print(round(df.drop(labels=['Entry', 'Occupancy'], axis=1).describe(), 1))
```

```python
sns.pairplot(data=df, vars=['Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio'], hue='Occupancy')
```

```python
ratios = [0.25, 1/3, 0.5]
sets = []

for i in range(0, len(ratios)):
    ratio = ratios[i]
    print(ratio)
    X_Train, X_Test, Y_Train, Y_Test = train_test_split(
        df[['Temperature','Humidity','Light','CO2','HumidityRatio']],
        df['Occupancy'],
        random_state=123,
        test_size=ratio
    )
    sets.append(tuple((X_Train, X_Test, Y_Train, Y_Test)))
```

```python
for i in range(0, len(sets)):
    X_Train, X_Test, Y_Train, Y_Test = sets[i]
    print(f"Size of X_Test {i}: {len(X_Test)}")
    print(f"Size of Y_Test {i}: {len(Y_Test)}")
```

```python
# 3 Cluster-Sizes of 3, 7 and 9
k = [3,7,9]
knn_clusters = []

for size in k:
    print('Generating Clusters')
    knn_clusters.append(KNeighborsClassifier(n_neighbors=size))

for cluster in knn_clusters:
    print('Fitting Training Data to Cluster')
    for i in range(0, len(sets)):
        X_Train, X_Test, Y_Train, Y_Test = sets[i]
        cluster.fit(X_Train, Y_Train)
```

```python
predictions = []
for cluster in knn_clusters:
    for i in range(0, len(sets)):
        X_Train, X_Test, Y_Train, Y_Test = sets[i]
        predictions.append(cluster.predict(X_Test))

for prediction in predictions:
    print(prediction)
```

```python
matrices = []
for prediction in predictions:
    for i in range(0, len(sets)):
        X_Train, X_Test, Y_Train, Y_Test = sets[i]
        if len(prediction) == len(Y_Test):
            cross = pd.crosstab(Y_Test, prediction, rownames=['Actual'], colnames=['Predicted'])
            matrices.append(cross)

for matrix in matrices:
    matrix
```
